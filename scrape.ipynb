```python
import requests
import json
```

```python
url = "https://statsapi.mlb.com/api/v1/schedule/postseason?sportId=1&season=2017"
```

```python
req = requests.get(url)
```

```python
obj = json.loads(req.text)
```

```python
obj['dates'][0]
```

```python
obj['dates'][0]['games'][0].keys()
```

```python
obj['dates'][2]['games'][0]['venue']
```

```python
astros = []
for gameday in obj['dates']:
    for game in gameday['games']:
        if game['venue']['id'] == 2392:
            astros.append(game)
```

```python
astros = list(filter(lambda x: x['games'][0]['venue']['id'] == 2392, obj['dates']))
```
